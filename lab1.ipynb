```python
import numpy as np
import csv
import matplotlib.pyplot as plt
```

```python

# зчитуємо дані з датасетів
train_file_path = 'lab_1_train.csv'
test_file_path = "lab_1_test.csv"

x_train_data = []
y_train_data = []
counter = 0
with open(train_file_path, 'r') as csvfile:
    csv_reader = csv.reader(csvfile)
    for row in csv_reader:
        if counter == 0:
            counter += 1
            continue
        x_train_data.append(float(row[1]))
        y_train_data.append(float(row[2]))

x_test_data = []
y_test_data = []
counter = 0
with open(test_file_path, 'r') as csvfile:
    csv_reader = csv.reader(csvfile)
    for row in csv_reader:
        if counter == 0:
            counter += 1
            continue
        x_test_data.append(float(row[1]))
        y_test_data.append(float(row[2]))

print(x_train_data)
print(y_train_data)
print(x_test_data)
print(y_test_data)
```

```python
#візуалізуємо дані з трейн датасету 
plt.scatter(x_train_data, y_train_data, s=4.5)
plt.show()
```

```python
beta0 = np.random.randn(1)
beta1 = np.random.randn(1)
print(beta0,beta1)

# коефіцієнт швидкості навчання
lr = 0.1

n_epochs = 1000


for epoch in range(n_epochs):
  
    # розраховуємо результуючий масив
    # на основі train-датасету
    yhat = beta0 + beta1 * x_train_data
    
    # обчислюємо відхилення нового результату від навчального
    error = (y_train_data - yhat)
    
    # обчислюємо градієнти
    a_grad = -error.mean()
    b_grad = -(x_train_data * error).mean()
    
    # оновлюємо коефіцієнти
    beta0 = beta0 - lr * a_grad
    beta1 = beta1 - lr * b_grad

    # обчислюємо функцію втрат
    loss = 0
    for i in range(len(x_train_data)):
        y_i = beta0 + beta1 * x_train_data[i]
        loss += ((y_i - y_train_data[i]) ** 2)
    loss /= len(x_train_data)

    print("beta0 =", beta0, "beta1 =", beta1, "loss function value =", loss)  
```

```python
def func(x): 
    return beta0 + beta1 * x

# виводимо на графік дані з обох датасетів
# а також лінію, до якої зійшовся розв'язок
x_values = np.linspace(0, 1.0, 100)
y_values = func(x_values)

plt.scatter(x_train_data, y_train_data, s=4.5)

plt.scatter(x_test_data, y_test_data, s=4.5)

plt.plot(x_values, y_values)
plt.xlabel('x')
plt.ylabel('y')

plt.show()
```

```python
#обчислюємо функцію втрат на тестових даних
loss = 0
for i in range(len(x_test_data)):
    y_i = beta0 + beta1 * x_test_data[i]
    loss += ((y_i - y_test_data[i]) ** 2)
loss /= len(x_test_data)

print("loss function value =", loss)
```
